```python
import pandas as pd
import re

# Load the CSV file into a DataFrame
file_path = 'question_answer.csv'  # Replace with your CSV file path
df = pd.read_csv(file_path)

# Define a function to clean text by removing special symbols
def clean_text(text):
    if isinstance(text, str):  # Ensure the input is a string
        return re.sub(r'[^A-Za-z0-9\s]', '', text)  # Keep alphanumeric and spaces only
    return text

# Apply the cleaning function to the relevant columns
df['Question'] = df['Question'].apply(clean_text)
df['Answer'] = df['Answer'].apply(clean_text)

# Save the cleaned data back to a CSV
output_path = 'cleaned_QA_dataset.csv'  # Define output file name
df.to_csv(output_path, index=False)

print(f"Cleaned data saved to {output_path}")
```
